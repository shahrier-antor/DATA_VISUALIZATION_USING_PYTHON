```python
def unique_val_count(data, column):
    
    #counting the unique falue
    df_count = data[column].value_counts()
    
    #Taking into dataframe
    df_count = pd.DataFrame(df_count)
    
    #Resettng index    
    df_count = df_count.reset_index()
    
    #Renaming columns 
    df_count = df_count.rename(columns={'index':column, column:'No_of_cars'})
    
    #Introducing new column for %
    df_count['% of cars'] = np.round((df_count['No_of_cars']/df_count['No_of_cars'].sum()*100), 2)
    
    return df_count
```

```python
    model_count = unique_val_count(df, 'model')
    year_count = unique_val_count(df, 'year')
    transmission_count = unique_val_count(df, 'transmission')
    fuelType_count = unique_val_count(df, 'fuelType')
```

```python
def barplot(data, column_x, color, rotation, yticks):
    
    #  barplot using seaborn
    sns.barplot(x=column_x, y='% of cars', data=data, color=color, alpha=0.75)
    
    # title of plot
    plt.title("Number of cars present for each " + column_x)
    
    # lebel for the x and y axis
    plt.xlabel(column_x)
    plt.ylabel("Percent of cars (%)")
    
    # rotation the xticks
    plt.xticks(rotation=rotation)
    
    # range for the yticks
    plt.yticks(yticks)
```

```python
    #sns.set_context('paper')

    plt.figure(figsize=(15,10))

    #(1,1)
    plt.subplot(2,2,1)
    barplot(model_count, 'model', 'blue', 90, np.arange(0,51,10))

    #(1,2)
    plt.subplot(2,2,2)
    barplot(year_count, 'year', 'orange', 90, np.arange(0,51,10))

    #(2,1)
    plt.subplot(2,2,3)
    barplot(transmission_count, 'transmission', 'green', 0, np.arange(0,71,10))

    #(2,2)
    plt.subplot(2,2,4)
    barplot(fuelType_count, 'fuelType', 'red', 0, np.arange(0,71,10))

    #title for all the plots
    plt.suptitle("Categorical variables in the SKODA used cars dataset")

    # individual plots  are separated
    plt.tight_layout()

    # display all the plots
    plt.show()
```

# Scatterplot and Trendline

```python
plt.figure(figsize=(20, 20))
plt.subplot(3, 1, 1)
plt.title("Scatterplot of cars present for each mileae against price")
sns.scatterplot(x='mileage', y='price', data = df, hue='transmission')


plt.subplot(3, 1, 2)
plt.title("Scatterplot of cars present for each mpg against price")
sns.regplot(x='mpg', y='price', data = df,scatter_kws= {'color': 'g', 'edgecolor': 'white'})
#scatter_kws is a part of matplotlib and color, edgecolor are dictionary

plt.subplot(3, 1, 3)
plt.title("Scatterplot of cars present for each engineSize against price")
sns.regplot(x='engineSize', y='price', data = df, line_kws= {'color': 'r'}) #line_kws is a part of matplotlib

plt.suptitle("ScatterPlot and trend line in the SKODA used cars dataset")

plt.tight_layout()
plt.show()
```

# Histogram 

```python
sns.set_context('talk')

plt.figure(figsize=(25, 20))

plt.subplot(3, 1, 1)
plt.title("Histogram of cars present for each mpg for default binwidth")
sns.histplot(data=df, x='mpg')#by default binwidth
plt.grid()

plt.subplot(3, 1, 2)
plt.title("Histogram of cars present for each mpg for binwidth 3")
sns.histplot(data=df, x='mpg', binwidth = 3, color='green', hue='transmission')
plt.xticks(np.arange(20, 90, 3))
plt.grid()

plt.subplot(3, 1, 3)
plt.title("Histogram of cars present for each mpg for binwidth 3 and fill false orange color")
sns.histplot(data=df, x='mpg', binwidth = 3, color='orange', fill=False)
plt.xticks(np.arange(20, 90, 3))
plt.grid()

plt.suptitle("Histogram in the SKODA used cars dataset")

plt.tight_layout()
plt.show()
```

# Distribution plot

```python
plt.figure(figsize=(20, 17))

plt.subplot(3, 1, 1)
plt.title("Distribution plot for mpg vs count")
sns.histplot(data=df, x='mpg', binwidth = 1,  kde = True)
plt.xticks(np.arange(20, 90, 5))

plt.subplot(3, 1, 2)
plt.title("Distribution plot for mpg vs count with transmission")
sns.histplot(data=df, x='mpg', binwidth = 1,  kde = True, hue='transmission')
plt.xticks(np.arange(20, 90, 5))

plt.suptitle("Distribution plot for mpg of Scoda cars")

plt.tight_layout()
plt.show()
```
